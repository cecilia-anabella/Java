```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('winequality.csv')
df.head()
```

```python
# what is the quality distribution in this dataset?
plt.figure(figsize=(4,3))
sns.countplot(data=df, x='quality', palette='RdPu', hue='quality')
plt.show()
```

```python
# explore sulphates central tendency and variances across quality
plt.figure(figsize=(7,4))
sns.pointplot(data=df, x='quality', y='sulphates', linestyles='--',
              marker='X', markersize=15, mfc='steelblue', mec='skyblue', color='hotpink')
plt.grid(axis='y',linewidth=0.5, color='pink')
plt.show()
```

```python
# explore central tendency and variance of sulphate and volitaile acidity
# across the quality
plt.figure(figsize=(7,4))
sns.pointplot(data=df, x='quality', y='sulphates', linestyles='--',
              marker='X', markersize=15, mfc='steelblue', mec='skyblue', color='hotpink',label='sulphates')
sns.pointplot(data=df, x='quality', y='volatile acidity',
               marker='o', markersize=10, mfc='lightgreen', mec='gold', label='volatile aciditly')
plt,plt.ylabel('value')
plt.legend()
plt.show()
```

```python
# explore the range (min, max, mean) of sulphates across quality
# explore the range (min, max, mean) of sulphates across quality
plt.figure(figsize=(7,4))
sns.pointplot(data=df, x='quality', y='sulphates', errorbar=('pi',100),
              capsize=0.2,color='hotpink', linestyle='none', marker='D',
              mfc='blue', mec='skyblue', markersize=10)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(7,4))
plt.figure(figsize=(7,4))
sns.stripplot(data=df, x='quality', y='citric acid', alpha=0.1,color='yellow')
sns.pointplot(data=df, x='quality', y='citric acid',marker='_',
              capsize=0.2,color='hotpink', linestyles='none', marker='D',
              mfc='blue', mec='skyblue', markersize=10)
plt.grid()
plt.show()
```
